```python
#Reading Excel file
import pandas as pd
df=pd.read_excel('/content/Random Price Dataset Trial Assignment.xlsx')
df
```

```python
#date column is irrelvant in prediction and hence removing the unwanted column
df.drop(['Date'],axis=1,inplace=True)
```

```python
#Column 'Saudi_export' converted to float
df['Saudi_export']=pd.to_numeric(df['Saudi_export'],errors='coerce')
```

```python
df.select_dtypes(include=['object']).head(0)
```

```python
#Null value check
dfnull=df.isna().sum()
null_columns = dfnull[dfnull>0].index.tolist()
null_columns
for i in null_columns:
  print(i)
```

```python
#null values replaced with mean values
for i in null_columns:
  df[i]=df[i].fillna(df[i].mean())
```

```python
dfnull=df.isna().sum()
null_columns = dfnull[dfnull>0].index.tolist()
null_columns
```

```python
# There is no values reported in 'South_korea_import' column and hence removing the same
df.drop(['South_korea_import'],axis=1,inplace=True)
```

```python
dfnull=df.isna().sum()
null_columns = dfnull[dfnull>0].index.tolist()
null_columns
```

```python
dfnull=pd.DataFrame(df.isna().sum())
dfnull
```

```python
#Feature X and class label y classification,
X=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
#feature values are scaled using Minmax scalar
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_new=sc.fit_transform(X)
```

```python
#Model is trained
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,BaggingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR, NuSVR
from sklearn.ensemble import ExtraTreesRegressor,AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
lr=LinearRegression()
rf=RandomForestRegressor(n_estimators=150)
svr=SVR()
nusvr=NuSVR()
etr=ExtraTreesRegressor()
ab=AdaBoostRegressor()
gbr=GradientBoostingRegressor()
knn=KNeighborsRegressor(n_neighbors=5)
mlp=MLPRegressor(hidden_layer_sizes=(100, ))
bag=BaggingRegressor()
tree=DecisionTreeRegressor()
lst=[lr,rf,svr,etr,ab,nusvr,gbr,knn,mlp,bag,tree]
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
for i in lst:
  
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(i,":",'r2_score:',r2_score(y_test,y_pred)*100)
  print(i,":",'mean_absolute_error:',mean_absolute_error(y_test,y_pred))
  import numpy as np
  print(i,":",'root mean_squared_error:',np.sqrt(mean_squared_error(y_test,y_pred)))
```
