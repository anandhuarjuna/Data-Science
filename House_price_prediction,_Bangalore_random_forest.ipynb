```python
import pandas as pd
df=pd.read_csv('/content/Bengaluru_House_Data (1).csv')
df.head()
```

```python
df.isna().sum()
```

```python
df= df.dropna(subset=['location'])
df= df.dropna(subset=['size'])
df= df.dropna(subset=['society'])
df= df.dropna(subset=['bath'])
df= df.dropna(subset=['balcony'])
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.drop(['availability'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['area_type']=le.fit_transform(df['area_type'])
df['location']=le.fit_transform(df['location'])
df['size']=le.fit_transform(df['size'])
df['society']=le.fit_transform(df['society'])
```

```python
df
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import MinMaxScaler 
sc=MinMaxScaler()
X_new=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=.3,random_state=3)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
rf=RandomForestRegressor(n_estimators=150)
lr=LinearRegression()
ri=Ridge(alpha=1.0)  
la=Lasso(alpha=1.0)
dt=DecisionTreeRegressor()
svr=SVR()
lst=[rf,lr,ri,la,dt,svr]
from sklearn.metrics import r2_score
for i in lst:
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print('****************')
  print(i)
  print(r2_score(y_test,y_pred)*100)
rf.predict(sc.transform([[]]))

```
