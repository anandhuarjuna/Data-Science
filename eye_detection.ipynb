```python
import cv2
image=cv2.imread(r'D:\Computer Vision\eye_detection\Many People.jpg')
image
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```

```python
con_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
eye_cas=cv2.CascadeClassifier(r'D:\Computer Vision\eye_detection\haarcascade_eye1.xml')
eye_cas
```

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
eye=eye_cas.detectMultiScale(gray_img,scaleFactor=1.1,minNeighbors=2)
```

```python
eye
```

```python
for x,y,w,h in eye:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(255,0,0),2)
    plt.imshow(con_img)
```

```python
video=cv2.VideoCapture(r'D:\Computer Vision\Face_Detection\Multiple people.mp4')
while True:
    sucess,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    eye=eye_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    for x,y,w,h in eye:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
        cv2.imshow('Video1',frame)
        if cv2.waitKey(1) & 0XFF==27:
            break
video.release()
cv2.destroyAllWindows

    
```
